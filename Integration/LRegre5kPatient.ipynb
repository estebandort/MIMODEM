```python
# Confidence intervals
from sklearn.metrics import roc_auc_score
from math import sqrt

def roc_auc_ci(y_true, y_score, positive=1):
    AUC = roc_auc_score(y_true, y_score)
    N1 = sum(y_true == positive)
    N2 = sum(y_true != positive)
    Q1 = AUC / (2 - AUC)
    Q2 = 2*AUC**2 / (1 + AUC)
    SE_AUC = sqrt((AUC*(1 - AUC) + (N1 - 1)*(Q1 - AUC**2) + (N2 - 1)*(Q2 - AUC**2)) / (N1*N2))
    lower = AUC - 1.96*SE_AUC
    upper = AUC + 1.96*SE_AUC
    if lower < 0:
        lower = 0
    if upper > 1:
        upper = 1
    return (lower, upper)
```

```python
# Labels 
y_labels = glam['malignant_label'] 
# The model predictions are concatenated in X_predic
X_predic = pd.concat([end2end.loc[:, ['malignant_pred']], frcnn.loc[:, ['malignant_pred']], glam.loc[:, ['malignant_pred']], gmic.loc[:, ['malignant_pred']], nyu_single.loc[:, ['malignant_pred']]], axis=1)
# Each column is named according to the model
X_predic.columns = ['mal_pred_end2end', 'mal_pred_frcnn', 'mal_pred_glam', 'mal_pred_gmic', 'mal_pred_nyu_single']
```

```python
# prediction calculation by patient
Y = y_labels.to_numpy()
X = X_predic.to_numpy()
Xp = np.zeros((382, 5))
Yp = np.zeros((382, 1))
np.array(Y, dtype=bool)

for i in  range(382):
     Xp[i,:] =  (X[4*i,:]+X[4*i+1,:]+X[4*i+2,:]+X[4*i+3,:])/4
     Yp[i] = ( Y[4*i] or Y[4*i+1] or Y[4*i+2] or Y[4*i+3] ) 

X_predic_p = pd.DataFrame(Xp, columns = ['mal_pred_end2end', 'mal_pred_frcnn', 'mal_pred_glam', 'mal_pred_gmic', 'mal_pred_nyu_single'])
y_labels_p = pd.DataFrame(Yp, columns = ['malignant_label'])
```

```python
# Folds creation 
# Each fold has 76 patients - 38 cases / 38 controls
# Fold 1 
X_fold1 = pd.concat([X_predic_p.iloc[0:38,:], X_predic_p.iloc[191:229,:]], axis=0)
Y_fold1 = pd.concat([y_labels_p.iloc[0:38], y_labels_p.iloc[191:229]], axis=0)
# Fold 2 
X_fold2 = pd.concat([X_predic_p.iloc[38:76,:], X_predic_p.iloc[229:267,:]], axis=0)
Y_fold2 = pd.concat([y_labels_p.iloc[38:76], y_labels_p.iloc[229:267]], axis=0)
# Fold 3  
X_fold3 = pd.concat([X_predic_p.iloc[76:114,:], X_predic_p.iloc[267:305,:]], axis=0)
Y_fold3 = pd.concat([y_labels_p.iloc[76:114], y_labels_p.iloc[267:305]], axis=0)
# Fold 4 
X_fold4 = pd.concat([X_predic_p.iloc[114:152,:], X_predic_p.iloc[305:343,:]], axis=0)
Y_fold4 = pd.concat([y_labels_p.iloc[114:152], y_labels_p.iloc[305:343]], axis=0)
# Fold 5 
X_fold5 = pd.concat([X_predic_p.iloc[152:191,:], X_predic_p.iloc[343:382,:]], axis=0)
Y_fold5 = pd.concat([y_labels_p.iloc[152:191], y_labels_p.iloc[343:382]], axis=0)
```
